#### Kmeans

I will create 5 clusters, I do not have a strong argumentation for that but I think the  "Average standardised disposable income of private households" will have a strong impact on the way the categories are formed. And by creating 5 clusters the incomes will be split into 5.

````python
#setting up the KMEANS
kmeans = cluster.KMeans(n_clusters=5)
````

````python
np.random.seed(1234)
kmeanscls = kmeans.fit(nbh_joined1[columns_to_plot])
````

````python
kmeanscls.labels_
````

````python
nbh_joined1['kmeans'] = kmeanscls.labels_
````

````python
size_of_clusters = nbh_joined1.groupby('kmeans').size()
size_of_clusters
````

As you can see the 5th cluster only has 2 neighbourhoods within it. I probably could have done with 1 cluster less, but I will see what the characteristics of this cluster is in the further analysis.

````python
# I used the same steps as in the 6th lab to set up the facet grids to visualise the characteristics of each cluster
for_plot = nbh_joined1.set_index('kmeans')
for_plot = for_plot[columns_to_plot]
for_plot.head()
````

````python
for_plot = for_plot.stack()
for_plot.head(10000)
````

````python
for_plot = for_plot.reset_index()
for_plot.head()
````

````python
for_plot = for_plot.rename({'level_1': 'Characteristics', 0:'Values'}, axis =1)
for_plot.head()
````

````python
facets = sns.FacetGrid(data=for_plot, row='Characteristics', hue='kmeans', \
                  sharey=False, sharex=False, aspect=3)
_ = facets.map(sns.kdeplot, 'Values', shade= True).add_legend()
````

The first graph shows that the income levels are nicely divided per group. The other graphs show that there is overlap between each cluster. What is interesting is that cluster "4" is characterised by a high income level, but the energylabels are not high at all. The bottom two graphs show that the "Values" can go below 0 but that is actually not the case, see below I use the min function to check the lowest value.

````python
for_plot.min()
````

#### Creating the map with clusters

````python
# This code I also got from the 6th lab.

f, ax = plt.subplots(1, figsize=(9, 9))

nbh_joined1.plot(column='kmeans', categorical=True, legend=True, linewidth=1, ax=ax)
ax.set_axis_off()
plt.axis('equal')
plt.title('Grouping of neighbourhoods', font= 'georgia', fontsize = 20)
plt.show()
````

You can see that in the center of The Hague, cluster "3"  (gray) is dominant. This cluster characterised by low income, moderate score of energy labels and low score of the amount of solar panels/collectors installed. A possible explanation is that these households do not have a lot of disposable income, making transitioning towards more sustainable housing harder. 

#### Conduct PCA

````python
from sklearn.decomposition import PCA
````

````python
X= nbh_joined1[["Average standardised disposable income of private households",
                 "Score of energielabels in neighbourhood","Number of installations per capita"]]
pca = PCA(n_components=2)
pca.fit(X)
````

````python
print('Variance explained by the components: ', pca.explained_variance_ratio_)
````

As you can see the first component almost completely captures the variances of the data. So we can say that we can use this new component as a predictor. I understood from the class that "top" pca components are useful as new predictors.
